## **Data Cleaning**

````python
# Clean column names
df_crashes.columns = (
    df_crashes.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('/', '_')
)

# Handle missing values (Drop vs Impute)
missing_ratio = df_crashes.isnull().mean()
cols_to_drop = missing_ratio[missing_ratio > 0.6].index
df_crashes.drop(columns=cols_to_drop, inplace=True)

if 'collision_id' in df_crashes.columns:
    df_crashes.dropna(subset=['collision_id'], inplace=True)

# Impute numeric columns
num_cols = df_crashes.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    if df_crashes[col].isnull().sum() > 0:
        df_crashes[col] = df_crashes[col].fillna(df_crashes[col].median())

# Impute categorical columns
cat_cols = df_crashes.select_dtypes(include=['object']).columns
for col in cat_cols:
    if df_crashes[col].isnull().sum() > 0:
        df_crashes[col] = df_crashes[col].fillna('Unknown')

# Show summary
print(f"Dropped columns: {list(cols_to_drop)}")
print("Remaining missing values per column:")
print(df_crashes.isnull().sum().sort_values(ascending=False).head(10))
````

````python
import numpy as np

# -------------------------------
# Detect & Fix Outliers
# -------------------------------

# 1️⃣ IQR method for numeric columns
num_cols = df_crashes.select_dtypes(include=[np.number]).columns

for col in num_cols:
    # Skip columns with all zeros or only one unique value
    if df_crashes[col].nunique() <= 1:
        continue

    Q1 = df_crashes[col].quantile(0.25)
    Q3 = df_crashes[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Clip values outside the IQR range
    df_crashes[col] = df_crashes[col].clip(lower_bound, upper_bound)

# -------------------------------
# 2️⃣ Logical domain rules
# -------------------------------
# Injuries and fatalities should never be negative
for col in ['number_of_persons_injured', 'number_of_persons_killed',
            'number_of_pedestrians_injured', 'number_of_pedestrians_killed']:
    if col in df_crashes.columns:
        df_crashes[col] = df_crashes[col].clip(lower=0)

# Latitude must be between roughly 40 and 41 (for NYC)
if 'latitude' in df_crashes.columns:
    df_crashes.loc[(df_crashes['latitude'] < 40) | (df_crashes['latitude'] > 41), 'latitude'] = np.nan

# Longitude must be between roughly -75 and -73
if 'longitude' in df_crashes.columns:
    df_crashes.loc[(df_crashes['longitude'] < -75) | (df_crashes['longitude'] > -73), 'longitude'] = np.nan

# -------------------------------
# 3️⃣ Summary of changes
# -------------------------------
print("✅ Outlier detection and correction complete.")
print("Numeric columns checked:", list(num_cols))
print("Latitude/Longitude cleaned using domain rules.")
````

````python
# =============================================================================
# COMPLETE DATA PROCESSING: LOADING, DUPLICATE REMOVAL & STANDARDIZATION
# =============================================================================

import pandas as pd
import numpy as np
from datetime import datetime

# =============================================================================
# 1. LOAD THE DATASETS
# =============================================================================

print("LOADING NYC MOTOR VEHICLE COLLISIONS DATASETS")
print("=" * 60)

# Load crashes dataset
crashes_url = 'https://data.cityofnewyork.us/api/views/h9gi-nx95/rows.csv?accessType=download'
df_crashes = pd.read_csv(crashes_url, low_memory=False)

# Load persons dataset
persons_url = 'https://data.cityofnewyork.us/api/views/f55k-p6yu/rows.csv?accessType=download'
df_persons = pd.read_csv(persons_url, low_memory=False)

print(f"✓ Crashes dataset loaded: {df_crashes.shape}")
print(f"✓ Persons dataset loaded: {df_persons.shape}")

# =============================================================================
# 2. REMOVE DUPLICATES
# =============================================================================

def remove_duplicates(df_crashes, df_persons):
    """
    Remove duplicate records from both datasets
    """
    print("\n" + "=" * 60)
    print("REMOVING DUPLICATES")
    print("=" * 60)

    crashes_clean = df_crashes.copy()
    persons_clean = df_persons.copy()

    # Store original shapes
    original_crashes_shape = crashes_clean.shape
    original_persons_shape = persons_clean.shape

    # =========================================================================
    # 2.1 Remove Exact Duplicates
    # =========================================================================
    print("\n1. REMOVING EXACT DUPLICATES")
    print("-" * 30)

    # Remove exact duplicates from crashes dataset
    crashes_before = len(crashes_clean)
    crashes_clean = crashes_clean.drop_duplicates()
    crashes_after = len(crashes_clean)
    crashes_duplicates = crashes_before - crashes_after

    print(f"Crashes dataset:")
    print(f"  Before: {crashes_before:,} records")
    print(f"  After:  {crashes_after:,} records")
    print(f"  Removed: {crashes_duplicates:,} exact duplicates ({crashes_duplicates/crashes_before*100:.2f}%)")

    # Remove exact duplicates from persons dataset
    persons_before = len(persons_clean)
    persons_clean = persons_clean.drop_duplicates()
    persons_after = len(persons_clean)
    persons_duplicates = persons_before - persons_after

    print(f"Persons dataset:")
    print(f"  Before: {persons_before:,} records")
    print(f"  After:  {persons_after:,} records")
    print(f"  Removed: {persons_duplicates:,} exact duplicates ({persons_duplicates/persons_before*100:.2f}%)")

    # =========================================================================
    # 2.2 Remove Near-Duplicates Based on Key Columns
    # =========================================================================
    print("\n2. REMOVING NEAR-DUPLICATES")
    print("-" * 30)

    # For crashes dataset: Remove duplicates based on collision ID and key attributes
    if 'collision_id' in crashes_clean.columns:
        crashes_before = len(crashes_clean)

        # Define key columns for identifying near-duplicates in crashes
        crashes_key_cols = ['collision_id', 'crash_date', 'crash_time', 'borough', 'zip_code']
        available_crashes_keys = [col for col in crashes_key_cols if col in crashes_clean.columns]

        if available_crashes_keys:
            crashes_clean = crashes_clean.drop_duplicates(subset=available_crashes_keys, keep='first')
            crashes_after = len(crashes_clean)
            crashes_near_duplicates = crashes_before - crashes_after

            print(f"Crashes dataset (near-duplicates):")
            print(f"  Key columns used: {available_crashes_keys}")
            print(f"  Before: {crashes_before:,} records")
            print(f"  After:  {crashes_after:,} records")
            print(f"  Removed: {crashes_near_duplicates:,} near-duplicates")

    # For persons dataset: Remove duplicates based on person ID and collision ID
    persons_before = len(persons_clean)

    # Define key columns for identifying near-duplicates in persons
    persons_key_cols = ['COLLISION_ID', 'UNIQUE_ID', 'PERSON_ID', 'PERSON_TYPE', 'PERSON_AGE']
    available_persons_keys = [col for col in persons_key_cols if col in persons_clean.columns]

    if available_persons_keys:
        persons_clean = persons_clean.drop_duplicates(subset=available_persons_keys, keep='first')
        persons_after = len(persons_clean)
        persons_near_duplicates = persons_before - persons_after

        print(f"Persons dataset (near-duplicates):")
        print(f"  Key columns used: {available_persons_keys}")
        print(f"  Before: {persons_before:,} records")
        print(f"  After:  {persons_after:,} records")
        print(f"  Removed: {persons_near_duplicates:,} near-duplicates")

    # =========================================================================
    # 2.3 Summary Statistics
    # =========================================================================
    print("\n3. DUPLICATE REMOVAL SUMMARY")
    print("-" * 30)

    total_crashes_removed = original_crashes_shape[0] - len(crashes_clean)
    total_persons_removed = original_persons_shape[0] - len(persons_clean)

    print(f"Crashes Dataset:")
    print(f"  Original: {original_crashes_shape[0]:,} records")
    print(f"  Final:    {len(crashes_clean):,} records")
    print(f"  Total Removed: {total_crashes_removed:,} records ({total_crashes_removed/original_crashes_shape[0]*100:.2f}%)")

    print(f"Persons Dataset:")
    print(f"  Original: {original_persons_shape[0]:,} records")
    print(f"  Final:    {len(persons_clean):,} records")
    print(f"  Total Removed: {total_persons_removed:,} records ({total_persons_removed/original_persons_shape[0]*100:.2f}%)")

    return crashes_clean, persons_clean

# Execute duplicate removal
print("Starting duplicate removal process...")
crashes_deduped, persons_deduped = remove_duplicates(df_crashes, df_persons)

# =============================================================================
# 3. COMPREHENSIVE DATA STANDARDIZATION
# =============================================================================

def comprehensive_standardization(df_crashes, df_persons):
    """
    Complete standardization of NYC Motor Vehicle Collisions datasets
    Handles date/time, strings, categories, vehicle types, and contributing factors
    """

    print("\n" + "=" * 60)
    print("STARTING COMPREHENSIVE STANDARDIZATION")
    print("=" * 60)

    # Create copies to avoid modifying original dataframes
    crashes_clean = df_crashes.copy()
    persons_clean = df_persons.copy()

    # Display original shapes
    print(f"Original shapes - Crashes: {crashes_clean.shape}, Persons: {persons_clean.shape}")

    # Define clean_time function at the top level of the function
    def clean_time(time_str):
        if pd.isna(time_str) or str(time_str).strip().upper() in ['NAN', 'NONE', '']:
            return np.nan
        time_str = str(time_str).strip()
        if ':' in time_str:
            parts = time_str.split(':')
            if len(parts) >= 2:
                return f"{parts[0].zfill(2)}:{parts[1].zfill(2)}"
        return time_str

    # =========================================================================
    # 3.1 DATE/TIME STANDARDIZATION
    # =========================================================================
    print("\n1. DATE/TIME STANDARDIZATION")
    print("-" * 40)

    # Standardize crashes dataset date/time
    if 'crash_date' in crashes_clean.columns:
        crashes_clean['crash_date'] = pd.to_datetime(crashes_clean['crash_date'], errors='coerce')
        print(f"✓ Converted crash_date to datetime - Missing: {crashes_clean['crash_date'].isna().sum()}")

    if 'crash_time' in crashes_clean.columns:
        crashes_clean['crash_time'] = crashes_clean['crash_time'].apply(clean_time)
        print(f"✓ Standardized crash_time - Missing: {crashes_clean['crash_time'].isna().sum()}")

    # Create combined datetime for analysis
    if 'crash_date' in crashes_clean.columns and 'crash_time' in crashes_clean.columns:
        crashes_clean['crash_datetime'] = pd.to_datetime(
            crashes_clean['crash_date'].astype(str) + ' ' + crashes_clean['crash_time'].astype(str),
            errors='coerce'
        )
        crashes_clean['crash_year'] = crashes_clean['crash_datetime'].dt.year
        crashes_clean['crash_month'] = crashes_clean['crash_datetime'].dt.month
        crashes_clean['crash_day'] = crashes_clean['crash_datetime'].dt.day
        crashes_clean['crash_hour'] = crashes_clean['crash_datetime'].dt.hour
        crashes_clean['crash_dayofweek'] = crashes_clean['crash_datetime'].dt.day_name()
        print(f"✓ Created datetime features - Missing: {crashes_clean['crash_datetime'].isna().sum()}")

    # Standardize persons dataset date/time
    if 'CRASH_DATE' in persons_clean.columns:
        persons_clean['CRASH_DATE'] = pd.to_datetime(persons_clean['CRASH_DATE'], errors='coerce')
        print(f"✓ Converted persons CRASH_DATE to datetime")

    if 'CRASH_TIME' in persons_clean.columns:
        persons_clean['CRASH_TIME'] = persons_clean['CRASH_TIME'].apply(clean_time)
        print(f"✓ Standardized persons CRASH_TIME")

    # =========================================================================
    # 3.2 STRING STANDARDIZATION
    # =========================================================================
    print("\n2. STRING STANDARDIZATION")
    print("-" * 40)

    def standardize_string_column(series):
        """Standardize a string series"""
        return (
            series
            .astype(str)
            .str.strip()
            .str.upper()
            .replace({
                'NAN': 'UNKNOWN', 'NONE': 'UNKNOWN', '': 'UNKNOWN',
                'UNSPECIFIED': 'UNKNOWN', 'N/A': 'UNKNOWN', 'NOT APPLICABLE': 'UNKNOWN',
                'NONE.': 'UNKNOWN', 'UNAVAILABLE': 'UNKNOWN'
            })
        )

    # Standardize crashes dataset strings
    crashes_string_cols = [
        'borough', 'zip_code', 'on_street_name', 'cross_street_name',
        'contributing_factor_vehicle_1', 'contributing_factor_vehicle_2',
        'vehicle_type_code_1', 'vehicle_type_code_2'
    ]

    crashes_string_cols = [col for col in crashes_string_cols if col in crashes_clean.columns]
    for col in crashes_string_cols:
        crashes_clean[col] = standardize_string_column(crashes_clean[col])
        print(f"✓ Standardized {col}: {crashes_clean[col].nunique()} unique values")

    # Standardize persons dataset strings
    persons_string_cols = [
        'PERSON_TYPE', 'PERSON_INJURY', 'PERSON_AGE', 'PERSON_SEX',
        'EJECTION', 'EMOTIONAL_STATUS', 'BODILY_INJURY',
        'POSITION_IN_VEHICLE', 'SAFETY_EQUIPMENT',
        'PED_LOCATION', 'PED_ACTION', 'COMPLAINT', 'PED_ROLE',
        'CONTRIBUTING_FACTOR_1', 'CONTRIBUTING_FACTOR_2'
    ]

    persons_string_cols = [col for col in persons_string_cols if col in persons_clean.columns]
    for col in persons_string_cols:
        persons_clean[col] = standardize_string_column(persons_clean[col])
        print(f"✓ Standardized {col}: {persons_clean[col].nunique()} unique values")

    # =========================================================================
    # 3.3 CATEGORICAL STANDARDIZATION
    # =========================================================================
    print("\n3. CATEGORICAL STANDARDIZATION")
    print("-" * 40)

    # Standardize borough in crashes
    if 'borough' in crashes_clean.columns:
        borough_mapping = {
            'MANHATTAN': 'MANHATTAN', 'BROOKLYN': 'BROOKLYN', 'QUEENS': 'QUEENS',
            'BRONX': 'BRONX', 'STATEN ISLAND': 'STATEN ISLAND', 'STATEN IS': 'STATEN ISLAND'
        }
        crashes_clean['borough'] = crashes_clean['borough'].replace(borough_mapping)
        crashes_clean['borough'] = crashes_clean['borough'].fillna('UNKNOWN')
        borough_counts = crashes_clean['borough'].value_counts()
        print("✓ Borough distribution:")
        for borough, count in borough_counts.items():
            print(f"  {borough}: {count:,} records")

    # Standardize person type
    if 'PERSON_TYPE' in persons_clean.columns:
        person_type_mapping = {
            'PEDESTRIAN': 'PEDESTRIAN', 'PEDESTRIAN1': 'PEDESTRIAN', 'PEDESTRIAN2': 'PEDESTRIAN',
            'BICYCLIST': 'BICYCLIST', 'BICYCLIST1': 'BICYCLIST', 'BICYCLIST2': 'BICYCLIST',
            'DRIVER': 'DRIVER', 'DRIVER1': 'DRIVER', 'DRIVER2': 'DRIVER',
            'PASSENGER': 'PASSENGER', 'PASSENGER1': 'PASSENGER', 'PASSENGER2': 'PASSENGER',
            'IN-LINE SKATER': 'OTHER', 'SKATEBOARDER': 'OTHER'
        }
        persons_clean['PERSON_TYPE'] = persons_clean['PERSON_TYPE'].replace(person_type_mapping)
        persons_clean['PERSON_TYPE'] = persons_clean['PERSON_TYPE'].fillna('UNKNOWN')
        person_type_counts = persons_clean['PERSON_TYPE'].value_counts()
        print("✓ Person type distribution:")
        for ptype, count in person_type_counts.items():
            print(f"  {ptype}: {count:,} records")

    # Standardize person injury
    if 'PERSON_INJURY' in persons_clean.columns:
        injury_mapping = {
            'KILLED': 'FATAL', 'INJURED': 'INJURED', 'INCAPACITATING INJURY': 'INJURED',
            'NON-INCAPACITATING INJURY': 'INJURED', 'POSSIBLE INJURY': 'INJURED', 'UNINJURED': 'UNINJURED'
        }
        persons_clean['PERSON_INJURY'] = persons_clean['PERSON_INJURY'].replace(injury_mapping)
        persons_clean['PERSON_INJURY'] = persons_clean['PERSON_INJURY'].fillna('UNKNOWN')
        injury_counts = persons_clean['PERSON_INJURY'].value_counts()
        print("✓ Person injury distribution:")
        for injury, count in injury_counts.items():
            print(f"  {injury}: {count:,} records")

    # Standardize person sex
    if 'PERSON_SEX' in persons_clean.columns:
        sex_mapping = {
            'M': 'MALE', 'F': 'FEMALE', 'MALE': 'MALE', 'FEMALE': 'FEMALE'
        }
        persons_clean['PERSON_SEX'] = persons_clean['PERSON_SEX'].replace(sex_mapping)
        persons_clean['PERSON_SEX'] = persons_clean['PERSON_SEX'].fillna('UNKNOWN')
        sex_counts = persons_clean['PERSON_SEX'].value_counts()
        print("✓ Person sex distribution:")
        for sex, count in sex_counts.items():
            print(f"  {sex}: {count:,} records")

    # =========================================================================
    # 3.4 VEHICLE TYPE STANDARDIZATION
    # =========================================================================
    print("\n4. VEHICLE TYPE STANDARDIZATION")
    print("-" * 40)

    vehicle_type_mapping = {
        'PASSENGER VEHICLE': ['PASSENGER', 'SEDAN', 'CAR', 'AUTO', 'PASSENGER CAR', 'TAXI', 'CAB'],
        'SPORT UTILITY VEHICLE': ['SUV', 'SPORT UTILITY', 'SPORT UTILITY VEHICLE'],
        'BUS': ['BUS', 'SCHOOL BUS', 'TRANSIT BUS'],
        'TRUCK': ['TRUCK', 'PICKUP TRUCK', 'DUMP TRUCK', 'TRUCK TRACTOR', 'VAN'],
        'MOTORCYCLE': ['MOTORCYCLE', 'MOTORCYCLE SCOOTER', 'SCOOTER', 'MOPED'],
        'BICYCLE': ['BICYCLE', 'BIKE'],
        'AMBULANCE': ['AMBULANCE', 'RESCUE VEHICLE'],
        'FIRE TRUCK': ['FIRE TRUCK', 'FIRE ENGINE'],
        'POLICE VEHICLE': ['POLICE', 'POLICE VEHICLE', 'POLICE CAR'],
        'UNKNOWN': ['UNKNOWN', 'NAN', 'NONE', 'OTHER', 'UNSPECIFIED']
    }

    def standardize_vehicle_type(vehicle_type):
        if pd.isna(vehicle_type) or str(vehicle_type).strip().upper() in ['', 'NAN']:
            return 'UNKNOWN'
        vehicle_str = str(vehicle_type).strip().upper()
        for standardized_type, variants in vehicle_type_mapping.items():
            if any(variant in vehicle_str for variant in variants):
                return standardized_type
        return 'OTHER'

    vehicle_cols = ['vehicle_type_code_1', 'vehicle_type_code_2']
    for col in vehicle_cols:
        if col in crashes_clean.columns:
            crashes_clean[f'{col}_standardized'] = crashes_clean[col].apply(standardize_vehicle_type)
            unique_types = crashes_clean[f'{col}_standardized'].nunique()
            top_values = crashes_clean[f'{col}_standardized'].value_counts().head(3)
            print(f"✓ {col}: {unique_types} types, Top: {dict(top_values)}")

    # =========================================================================
    # 3.5 CONTRIBUTING FACTOR STANDARDIZATION
    # =========================================================================
    print("\n5. CONTRIBUTING FACTOR STANDARDIZATION")
    print("-" * 40)

    contributing_factor_mapping = {
        'DRIVER INATTENTION': ['INATTENTION', 'DISTRACTION', 'CELL PHONE', 'DISTRACTED'],
        'SPEEDING': ['SPEED', 'SPEEDING', 'EXCEEDING SPEED LIMIT'],
        'ALCOHOL INVOLVEMENT': ['ALCOHOL', 'DRINKING', 'INTOXICATION'],
        'FAILURE TO YIELD': ['FAILURE TO YIELD', 'YIELD'],
        'FOLLOWING TOO CLOSELY': ['FOLLOWING TOO CLOSELY', 'TAILGATING'],
        'UNSAFE LANE CHANGING': ['LANE CHANGING', 'IMPROPER LANE CHANGE'],
        'TRAFFIC CONTROL DISREGARDED': ['TRAFFIC CONTROL', 'SIGNAL', 'STOP SIGN', 'RED LIGHT'],
        'UNKNOWN': ['UNKNOWN', 'NAN', 'NONE', 'UNSPECIFIED', 'OTHER']
    }

    def standardize_contributing_factor(factor):
        if pd.isna(factor) or str(factor).strip().upper() in ['', 'NAN']:
            return 'UNKNOWN'
        factor_str = str(factor).strip().upper()
        for standardized_factor, variants in contributing_factor_mapping.items():
            if any(variant in factor_str for variant in variants):
                return standardized_factor
        return 'OTHER'

    factor_cols = ['contributing_factor_vehicle_1', 'contributing_factor_vehicle_2']
    for col in factor_cols:
        if col in crashes_clean.columns:
            crashes_clean[f'{col}_standardized'] = crashes_clean[col].apply(standardize_contributing_factor)
            unique_factors = crashes_clean[f'{col}_standardized'].nunique()
            top_values = crashes_clean[f'{col}_standardized'].value_counts().head(3)
            print(f"✓ {col}: {unique_factors} factors, Top: {dict(top_values)}")

    # =========================================================================
    # 3.6 FINAL DATA QUALITY SUMMARY
    # =========================================================================
    print("\n6. FINAL DATA QUALITY SUMMARY")
    print("-" * 40)

    print("CRASHES DATASET:")
    print(f"  Total records: {len(crashes_clean):,}")
    if 'crash_datetime' in crashes_clean.columns:
        print(f"  Date range: {crashes_clean['crash_datetime'].min()} to {crashes_clean['crash_datetime'].max()}")
    if 'borough' in crashes_clean.columns:
        borough_summary = crashes_clean['borough'].value_counts()
        print(f"  Boroughs: {len(borough_summary)} unique values")

    print("\nPERSONS DATASET:")
    print(f"  Total records: {len(persons_clean):,}")
    if 'PERSON_TYPE' in persons_clean.columns:
        person_type_summary = persons_clean['PERSON_TYPE'].value_counts()
        print(f"  Person types: {len(person_type_summary)} unique values")
    if 'PERSON_INJURY' in persons_clean.columns:
        injury_summary = persons_clean['PERSON_INJURY'].value_counts()
        print(f"  Injury types: {len(injury_summary)} unique values")

    print("\n" + "=" * 60)
    print("STANDARDIZATION COMPLETE!")
    print("=" * 60)

    return crashes_clean, persons_clean

# Execute comprehensive standardization
print("\nStarting comprehensive standardization process...")
crashes_final, persons_final = comprehensive_standardization(crashes_deduped, persons_deduped)

# =============================================================================
# 4. DISPLAY FINAL RESULTS
# =============================================================================

print("\n" + "=" * 60)
print("FINAL RESULTS")
print("=" * 60)

print("\nFINAL SAMPLE - Crashes Dataset:")
print(crashes_final[['crash_date', 'crash_time', 'borough', 'vehicle_type_code_1_standardized', 'contributing_factor_vehicle_1_standardized']].head(2))

print("\nFINAL SAMPLE - Persons Dataset:")
print(persons_final[['PERSON_TYPE', 'PERSON_INJURY', 'PERSON_SEX', 'PERSON_AGE']].head(2))

print(f"\nFINAL DATASET SHAPES:")
print(f"Crashes: {crashes_final.shape}")
print(f"Persons: {persons_final.shape}")

print(f"\nPROCESSING COMPLETE!")
print(f"✓ Data loaded successfully")
print(f"✓ Duplicates removed")
print(f"✓ All formats standardized")
print(f"✓ Data ready for analysis and visualization")
````

## Phase 1: Duplicate Removal

### 1.1 Exact Duplicate Removal
**Method**: Used `drop_duplicates()` without specifying columns to remove fully identical rows.

**Justification**:
- Fully identical records represent data entry errors or system duplicates
- Provide no additional analytical value
- Reduce computational load for processing

**Impact**:
- Removed exact duplicates from both datasets
- Maintained data integrity while eliminating redundancy

### 1.2 Near-Duplicate Removal
**Crashes Dataset Key Columns**: `collision_id`, `crash_date`, `crash_time`, `borough`, `zip_code`

**Justification**:
- These combinations should uniquely identify events/individuals in their respective contexts
- Prevents overcounting of the same incident or person
- Ensures each record represents a distinct entity

**Decision**: Used `keep='first'` to preserve the first occurrence of duplicate records.
